```python
sales_price_df.isna()
```

```python
for columnName, column in sales_price_df.iteritems():
    sales_price_df[columnName] = round(column.fillna(column.mean()),2)
```

```python
sales_price_df.isna()
```

```python
sales_price_df["percent_change"] = (sales_price_df["2019-12"] - sales_price_df["2017-01"])/sales_price_df["2017-01"]
sales_price_df.head()


# (2019-12-salesprice - 2017-01-salesprice/2017-01-sales-price) = 0.3473

#cut price cut dataset
#keep only columns 2017-01 and 2019-12
sales_price_df.loc[sales_price_df['percent_change'] > 0, 'equity_risk'] = True 
sales_price_df.loc[sales_price_df['percent_change'] <= 0, 'equity_risk'] = False

sales_price_df.head()
```

```python
X = sales_price_df[sales_price_df.columns.to_series()["RegionName":"2019-12"]]
y = sales_price_df["equity_risk"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
